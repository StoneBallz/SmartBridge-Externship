Required imports

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers
from tensorflow.keras.layers import Convolution2D,MaxPooling2D,Flatten,Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing import image
```

## Task 1 - Data Reading, Pre-processing and Augmentation

```python
datagen = ImageDataGenerator(
    rotation_range=20,    
    width_shift_range=0.2,  
    height_shift_range=0.2, 
    shear_range=0.2,        
    zoom_range=0.2,         
    horizontal_flip=True,   
    fill_mode='nearest'
)
```

```python
train = datagen.flow_from_directory('/kaggle/input/bird-species-classification/test_data/test_data',
                                      target_size=(120, 120),
                                      class_mode='categorical', 
                                      batch_size=8)
test = datagen.flow_from_directory('/kaggle/input/bird-species-classification/train_data/train_data',
                                    target_size=(120, 120),
                                      class_mode='categorical', 
                                      batch_size=8)
```

```python
train.image_shape
```

## Task 2 - Model Building

```python
model = Sequential()
model.add(Convolution2D(20,(3,3),activation='relu',input_shape=(120, 120, 3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(45,activation='relu'))
model.add(Dense(16,activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(train,batch_size=8,validation_data=test,epochs=10)
```

## Task 3 - Testing Data

```python
img = image.load_img('/kaggle/input/bird-species-classification/test_data/test_data/blasti/DSC_6396.jpg',target_size=(120,120))
img = image.img_to_array(img)
img = np.expand_dims(img,axis=0)
prediction = np.argmax(model.predict(img))
for i in train.class_indices:
    if train.class_indices[i] == prediction:
        print(i)
        break
```
