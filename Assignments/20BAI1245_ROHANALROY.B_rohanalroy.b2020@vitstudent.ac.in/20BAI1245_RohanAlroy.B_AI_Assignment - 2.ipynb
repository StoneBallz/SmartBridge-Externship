## TASK - 2
Build the ANN model with (input layer, min 3 hidden layers & output layer)


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Initializing the seq model
input_shape = (5,)
rohan_model = Sequential()
#  input layer
rohan_model.add(Dense(5,activation='relu', input_shape=input_shape))
#  1st hidden layer
rohan_model.add(Dense(64,activation='relu'))
#  2nd hidden layer
rohan_model.add(Dense(128,activation='relu'))
#  3rd hidden layer
rohan_model.add(Dense(64,activation='relu'))
# output layer
rohan_model.add(Dense(5,activation='softmax'))

```

```python
rohan_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
rohan_model.summary()
```

```python
from tensorflow.keras.utils import to_categorical

# Convert labels to categorical format
ytrain_categorical = to_categorical(ytrain)

```

```python
ytrain_categorical
```

```python
xtrain
```

### TRAINING

```python
#training the model with xtrain and ytrain_categorical
rohan_model.fit(xtrain,ytrain_categorical,epochs=30,batch_size = 5)
```

Training with the scaled data

```python
# Initializing the seq model
input_shape = (5,)
rohan_models = Sequential()
#  input layer
rohan_models.add(Dense(5,activation='relu', input_shape=input_shape))
#  1st hidden layer
rohan_models.add(Dense(64,activation='relu'))
#  2nd hidden layer
rohan_models.add(Dense(128,activation='relu'))
#  3rd hidden layer
rohan_models.add(Dense(64,activation='relu'))
# output layer
rohan_models.add(Dense(5,activation='softmax'))

rohan_models.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
rohan_models.fit(xtrains,ytrain_categorical,epochs=30,batch_size = 5)
```
